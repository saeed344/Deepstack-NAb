```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import scale
from tensorflow.keras.layers import GRU
from keras.layers import Dense,Dropout
from keras.models import Sequential
from keras.layers import Flatten
from keras import regularizers
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import scale
from keras.layers import Conv1D,AveragePooling1D, MaxPooling1D
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier,GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
import lightgbm as lgb
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
import math,os
from sklearn.metrics import roc_curve, auc


def categorical_probas_to_classes(p):
    p = np.atleast_2d(p)
    print(p)
    return np.argmax(p, axis=1)

def to_categorical(y, nb_classes=None):
    y = np.array(y, dtype='int')
    if not nb_classes:
        nb_classes = np.max(y)+1
    Y = np.zeros((len(y), nb_classes))
    for i in range(len(y)):
        Y[i, y[i]] = 1
    return Y

def calculate_performace(test_num, pred_y,  labels):
    tp =0
    fp = 0
    tn = 0
    fn = 0
    for index in range(test_num):
        if labels[index] ==1:
            if labels[index] == pred_y[index]:
                tp = tp +1
            else:
                fn = fn + 1
        else:
            if labels[index] == pred_y[index]:
                tn = tn +1
            else:
                fp = fp + 1

    acc = float(tp + tn)/test_num
    precision = float(tp)/(tp+ fp + 1e-06)
    npv = float(tn)/(tn + fn + 1e-06)
    sensitivity = float(tp)/ (tp + fn + 1e-06)
    specificity = float(tn)/(tn + fp + 1e-06)
    mcc = float(tp*tn-fp*fn)/(math.sqrt((tp+fp)*(tp+fn)*(tn+fp)*(tn+fn)) + 1e-06)
    f1=float(tp*2)/(tp*2+fp+fn+1e-06)
    return acc, precision,npv, sensitivity, specificity, mcc, f1

def get_shuffle(dataset,label):
    index = [i for i in range(len(label))]
    np.random.shuffle(index)
    dataset = dataset[index]
    label = label[index]
    return dataset,label
def get_GRU_model(input_dim,out_dim):
    model = Sequential()
    model.add(GRU(256,return_sequences=True,input_shape=(1,input_dim)))
    model.add(Dropout(0.05))
    model.add(GRU(256,return_sequences=True))
    model.add(Dropout(0.05))
    model.add(Flatten())
    model.add(Dense(128, activation = 'relu',name="Dense_128"))
    model.add(Dropout(0.05))
    model.add(Dense(2, activation = 'softmax',name="Dense_2"))
    model.compile(loss = 'binary_crossentropy', optimizer = 'Adam', metrics =['accuracy'])#rmsprop
    return model

def get_DNN_model(input_dim,out_dim):
    model = Sequential()
    model.add(Dense(int(128), activation='relu',  name='Dense__1',input_shape=(1,input_dim)))#init='glorot_normal',
    model.add(Dropout(0.05))
    model.add(Dense(int(128), activation='relu',  name='Dense__2'))#init='glorot_normal',
    model.add(Dropout(0.05))
    model.add(Flatten())
    model.add(Dense(int(64), activation='relu',  name='Dense_3'))#init='glorot_normal',
    model.add(Dropout(0.05))
    model.add(Dense(2, activation='softmax', name='output'))
    model.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics =['accuracy'])#rmsprop
    return model
def get_CNN_model(input_dim,out_dim):
    model = Sequential()
    model.add(Conv1D(filters = 64, kernel_size = 3, padding = 'same', activation= 'relu'))
    model.add(MaxPooling1D(pool_size=2,strides=1,padding="same"))
    model.add(Conv1D(filters = 64, kernel_size =  3, padding = 'same', activation= 'relu'))
    model.add(AveragePooling1D(pool_size=2,strides=1,padding="same"))
    model.add(Flatten())
    model.add(Dense(int(128), activation = 'relu'))
    model.add(Dense(out_dim, activation = 'softmax',name="Dense_2"))
    model.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics =['accuracy'])
    return model

def get_stacking(clf, x_train, y_train, x_test, num_class,n_folds=5):
    kf = KFold(n_splits=n_folds)
    second_level_train_set=[]
    second_level_train_CF = []
    test_nfolds_set_CF = []
    test_nfolds_set=[]
    for i,(train_index, test_index) in enumerate(kf.split(x_train)):
        x_tra, y_tra = x_train[train_index], y_train[train_index]
        x_tst, y_tst =  x_train[test_index], y_train[test_index]

        clf.fit(x_tra, y_tra,epochs=30)

        second_level_train_ = clf.predict(x_tst)[:,1]
        second_level_train_set.append(second_level_train_)

        second_level_train_predCF_ = (second_level_train_ >= 0.5).astype(int)#clf.predict(x_tst)[:,1]
        print('the predicat class shape',np.shape(second_level_train_predCF_))
        # second_level_train_CF_ = categorical_probas_to_classes(second_level_train_predCF)
        second_level_train_CF.append(second_level_train_predCF_)

        test_nfolds= clf.predict(x_test)[:,1]
        test_nfolds_set.append(test_nfolds)

        test_nfolds_preCF_= (test_nfolds >= 0.5).astype(int)
        # test_nfolds_set_CF = categorical_probas_to_classes(test_nfolds_preCF)
        test_nfolds_set_CF.append(test_nfolds_preCF_)

    return train_second_level,test_second_level,train_second_level_CF, test_second_level_CF

def get_stacking2(clf, x_train, y_train, x_test, num_class,n_folds=5):
    kf = KFold(n_splits=n_folds)
    second_level_train_set=[]
    test_nfolds_set=[]
    second_level_train_set_CF=[]
    test_nfolds_set_CF=[]
    for i,(train_index, test_index) in enumerate(kf.split(x_train)):
        x_tra, y_tra = x_train[train_index], y_train[train_index]
        x_tst, y_tst =  x_train[test_index], y_train[test_index]
        clf.fit(x_tra, y_tra)
        second_level_train_ = clf.predict_proba(x_tst)[:,1]
        second_level_train_set.append(second_level_train_)

        second_level_train_predCF = clf.predict(x_tst)




def get_first1_level(train_x, train_y, test_x,num_class):

    train_sets = []
    test_sets = []
    train_sets_CF = []
    test_sets_CF = []
    for clf in [model1,model2,model3]:
        train_set, test_set,train_set_CF, test_set_CF = get_stacking(clf, train_x, train_y, test_x,num_class)
        train_sets.append(train_set)
        test_sets.append(test_set)

        train_sets_CF.append(train_set_CF)
        test_sets_CF.append(test_set_CF)
    print("Shape of meta_test1_CF:", np.shape(meta_train1_CF))
    print("Shape of meta_test1_CF:", np.shape(meta_test1_CF))
    print("Shape of meta_train1:", np.shape(meta_train1))
    print("Shape of meta_test1:", np.shape(meta_test1))
    return meta_train1,meta_test1,meta_train1_CF,meta_test1_CF

def get_first2_level(train_x, train_y, test_x,num_class):
    model3 = MLPClassifier()
    model4 = SVC(probability=True,kernel='rbf')
    model5 = RandomForestClassifier()
    model6 = ExtraTreesClassifier()
    model7 = XGBClassifier()
    train_sets = []
    test_sets = []
    train_sets_CF = []
    test_sets_CF = []
    input_dim=np.shape(train_x)[1]
    for clf in [model3,model4,model5,model6,model7]:
        # train_x=np.reshape(train_x,(-1,input_dim,1))
        # test_x=np.reshape(test_x,(-1,input_dim,1))
        train_x=np.reshape(train_x,(-1,input_dim))
        test_x=np.reshape(test_x,(-1,input_dim))
        train_y=np.reshape(y_train2,(-1,1))
        train_set, test_set, train_set_CF, test_set_CF = get_stacking2(clf, train_x, train_y, test_x,num_class)
        train_sets.append(train_set)
        test_sets.append(test_set)
        train_sets_CF.append(train_set_CF)
        test_sets_CF.append(test_set_CF)

    print("Shape of meta_train2_CF:", np.shape(meta_train2_CF))
    print("Shape of meta_test1_CF:", np.shape(meta_test2_CF))
    print("Shape of meta_train1:", np.shape(meta_train2))
    print("Shape of meta_test1:", np.shape(meta_test2))
    return meta_train2,meta_test2,meta_train2_CF,meta_test2_CF

def get_second_level(train_dim,train_label,test_dim,num_class):
    meta_train1,meta_test1, meta_train1_CF,meta_test1_CF =get_first1_level(train_dim,train_label,test_dim,num_class)
    [aa,bb,cc]=np.shape(X_train)
    [aa1,bb1,cc1]=np.shape(X_test)
    meta_train1=np.reshape(meta_train1,(aa,3))
    meta_test1=np.reshape(meta_test1,(aa1,3))

    meta_train1_CF=np.reshape(meta_train1_CF,(aa,3))
    meta_test1_CF=np.reshape(meta_test1_CF,(aa1,3))
    meta_train2,meta_test2,meta_train2_CF,meta_test2_CF=get_first2_level(meta_train1,train_label,meta_test1,num_class)
  

    print("Shape of meta_train_fusion_CF_PF:", np.shape(meta_train_fusion_CF_PF))
    print("Shape of meta_test_fusion_CF_PF:", np.shape(meta_test_fusion_CF_PF))
    print("Shape of meta_test1:", np.shape(meta_test1))
    print("Shape of meta_test2:", np.shape(meta_test2))

    print("Shape of meta_train_fusion:", np.shape(meta_train_fusion))


    hist=best_svm_model.fit(meta_train_fusion,y_train2)
    pre_score=best_svm_model.predict_proba(meta_test_fusion)
    return meta_train_fusion,meta_test_fusion,pre_score

################################################################################
################################################################################


path = "/content/drive/MyDrive/Watashara_Projects/8-Dengue/"
data_ = pd.read_csv(path + 'Feature_Selection/GLasso_feat.csv')

data_np = np.array(data_)
data_scaled = (data_np[:, :])
print(np.shape(data_np))

label1=np.ones((554 ,1))#Value can be changed
label2=np.zeros((554 ,1))
labels=np.append(label1,label2)
y = labels

X_Train,X_Test,y_Train,y_Test = train_test_split(data_scaled, labels, test_size=0.2, random_state=42)
X = X_Train
y = y_Train



X = np.array(X, dtype=float)
# y = label


X,y=get_shuffle(X,y)
[sample_num,input_dim]=np.shape(X)
print(sample_num,input_dim)

folder_name = 'deepstacking_SVM_2'
if not os.path.exists(path + (f'Results/{folder_name}/')):
    os.makedirs(path + f'Results/{folder_name}/',exist_ok=True)

out_dim=2
num_class=2
n=input_dim

sepscores = []
ytest=np.ones((1,2))*0.5
yscore=np.ones((1,2))*0.5
meta_train_feat = np.ones((1,8))
meta_test_feat = np.ones((1,8))
skf= StratifiedKFold(n_splits=5)

for train, test in skf.split(X,y):
    X_train = np.reshape(X[train], (-1, 1, input_dim))
    X_test = np.reshape(X[test], (-1, 1, input_dim))
    y_train = to_categorical(y[train])  # generate the resonable results
    y_train2 = y[train]
    # print(np.shape(y_train2))
    y_test = to_categorical(y[test])  # generate the test
    [aa, bb, cc] = np.shape(X_train)
    input_dim2 = aa

    meta_train2, meta_test2, y_score = get_second_level(X_train, y_train, X_test, num_class)


    # Concatenate safely
    yscore = np.vstack((yscore, y_score))
    meta_train_feat = np.vstack((meta_train_feat, meta_train2))
    meta_test_feat = np.vstack((meta_test_feat, meta_test2))

    y_test=to_categorical(y[test])
    ytest=np.vstack((ytest,y_test))
    fpr, tpr, _ = roc_curve(y_test[:,0], y_score[:,0])
    roc_auc = auc(fpr, tpr)
    y_class= categorical_probas_to_classes(y_score)
    y_test_tmp=y[test]
    acc, precision,npv, sensitivity, specificity, mcc,f1 = calculate_performace(len(y_class), y_class, y_test_tmp)
    sepscores.append([acc, sensitivity, specificity, mcc,f1,roc_auc])
    print('Resultlist: acc=%f,sensitivity=%f,specificity=%f,mcc=%f,f1=%f,roc_auc=%f'
          % (acc,  sensitivity, specificity, mcc,f1, roc_auc))
    hist=[]
    cv_clf=[]
scores=np.array(sepscores)

print("acc=%.2f%% (+/- %.2f%%)" % (np.mean(scores, axis=0)[0]*100,np.std(scores, axis=0)[0]*100))
print("sensitivity=%.2f%% (+/- %.2f%%)" % (np.mean(scores, axis=0)[1]*100,np.std(scores, axis=0)[1]*100))
print("specificity=%.2f%% (+/- %.2f%%)" % (np.mean(scores, axis=0)[2]*100,np.std(scores, axis=0)[2]*100))
print("mcc=%.2f%% (+/- %.2f%%)" % (np.mean(scores, axis=0)[3]*100,np.std(scores, axis=0)[3]*100))
print("f1=%.2f%% (+/- %.2f%%)" % (np.mean(scores, axis=0)[4]*100,np.std(scores, axis=0)[4]*100))
print("roc_auc=%.2f%% (+/- %.2f%%)" % (np.mean(scores, axis=0)[5]*100,np.std(scores, axis=0)[5]*100))

result1=np.mean(scores,axis=0)
H1=result1.tolist()
sepscores.append(H1)
result=sepscores[2]

print('Results: ACC=%f,Se=%f,Sp=%f,Mcc=%f,F1=%f,ROC_AUC=%f'
      % (result[0],result[1],result[2],result[3],
         result[4],result[5]))
result1=np.mean(scores,axis=0)
H1=result1.tolist()
sepscores.append(H1)
result=sepscores
row=yscore.shape[0]
yscore=yscore[np.array(range(1,row)),:]
yscore_sum = pd.DataFrame(data=yscore)

row=meta_train_feat.shape[0]
meta_train_feat=meta_train_feat[np.array(range(1,row)),:]
meta_train_feat_sum = pd.DataFrame(data=meta_train_feat).to_csv(path + f'Results/{folder_name}/meta_train_feat.csv')

row=meta_test_feat.shape[0]
meta_test_feat=meta_test_feat[np.array(range(1,row)),:]
meta_test_feat_sum = pd.DataFrame(data=meta_test_feat).to_csv(path + f'Results/{folder_name}/meta_test_feat.csv')

ytest=ytest[np.array(range(1,row)),:]
ytest_sum = pd.DataFrame(data=ytest)
data_csv_zhibiao = pd.DataFrame(data=result)

yscore_sum.to_csv(path + f'Results/{folder_name}/yscore_DeepStack.csv')

ytest_sum.to_csv(path + f'Results/{folder_name}/ytest_DeepStack.csv')

data_csv_zhibiao.to_csv(path + f'Results/{folder_name}/Results_Deepstack.csv')
```
